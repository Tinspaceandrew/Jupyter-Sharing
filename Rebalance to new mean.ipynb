```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
sns.set(rc={'figure.figsize':(11.7,8.27)})  ##set graph plot size
```

```python
path="/Users/andrewjmorrison/Y10DA.csv"
df = pd.read_csv(path)# Dataset is now stored in a Pandas Dataframe
##df=pd.read_csv("/Users/andrewjmorrison/Y10DA.csv")
```

```python
df.columns=['Name','Surname','Bio','Chem','Phys']
df.head()
```

```python
df.describe()
```

```python
def mean(col_name):
  total=df[col_name].sum()
  count_row = df.shape[0]  # gives number of row count
  mean_num=total/(count_row)
  return(mean_num)
```

##Here we call the mean function inside the rebalance fn, make some columns to find z number, rebalance, then delete all columns except z number

```python
nEw_mean=65    ##set your new mean here
nEw_stdDEV=15   ## set new standard deviation here

def reBalance(nAme):
  x=mean(nAme)

  
  ##myLabel=str(input('Name for rebalanced column'))
  ##newLabel=myLabel+' rebal'
  newLabel=nAme+' rebal'
  df['residual']=df[nAme]-x
  df['residual^2']=(df[nAme]-x)**2
  variance=mean('residual^2')
  variance=round(variance,1)
  standev=variance**0.5
  standev=round(standev,1)
  df['z_number']=df['residual']/standev
  df[newLabel]=round((df['z_number']*nEw_stdDEV)+nEw_mean)
  del df['residual^2']
  del df['z_number']
  del df['residual']


reBalance('Bio')
reBalance('Chem')
reBalance('Phys')

del df['Bio']
del df['Chem']
del df['Phys']

df['Bio rebal'].where(df['Bio rebal'] <= 100, 100, inplace=True)
df['Phys rebal'].where(df['Phys rebal'] <= 100, 100, inplace=True)
df['Chem rebal'].where(df['Chem rebal'] <= 100, 100, inplace=True)

df.head()
 
```

```python
df.describe()
```

```python

df['Bio_resid']=df['Bio rebal']-nEw_mean
df['Chem_resid']=df['Chem rebal']-nEw_mean
df['Phys_resid']=df['Phys rebal']-nEw_mean
df=df[['Name','Surname','Bio rebal','Bio_resid','Chem rebal','Chem_resid','Phys rebal','Phys_resid',]]  ##change column order
df['Rebal Avg']=round((df['Bio rebal']+df['Chem rebal']+df['Phys rebal'])/3)
```
